```python
from sklearn.model_selection import train_test_split
# 题目一 使用C4.5决策树完成肿瘤预测（50分）： 
# 特征:nancies:妊娠,Glucose:血糖,BloodPressure:血压,SkinThickness:皮肤厚度,Insulin:胰岛素,BMI:身体质量指数,DiabetesPedigreeFunction:糖尿病谱系功能,Age:年龄
# 标签:Outcome:结局
# 1、读取附件diabetes.csv中数据 （5分）
import pandas as pd
import numpy as  py
a1=pd.read_csv("E:\\python代码\\zk\\zk3\\data\\diabetes.csv")
print("读取的附件diabetes.csv中数据 \n",a1)
# 2、查看数据的详细信息,并统计数据量（5分）
print("查看数据的详细信息 \n",a1.info)
print("统计数据量\n",a1.shape)
# 3、提取特征变量和目标变量（5分）
x=a1.iloc[:,:-1]
y=a1['Outcome']
# 4、划分训练集和测试集（5分）
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=42)
# 5、将用于分类的标签生成子数据集（5分）

# 6、创建决策树模型（5分）
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
# 7、使用交叉验证对模型进行调参（5分）
from sklearn.model_selection import GridSearchCV
parameters = {'max_depth':[1,2,3], 'min_samples_split':[2,3],'min_samples_leaf':[1,2]}
ty1 = GridSearchCV(clf,parameters,scoring='accuracy',cv=5)
print('交叉验证\n',ty1)

# 8、使用网格搜索对模型进行调参,输出最佳参数组合和对应的得分（5 分）
ty1.fit(X_train,y_train)
zy=ty1.best_params_
print('最佳参数组',zy)
zz=ty1.best_score_
print('对应的得分',zz)
# 9、打印出预测的结果（5分）
t9=ty1.predict(X_test)
print('预测的结果\n',t9)
# 10、绘制ROC曲线来评估模型预测的效果（5 分）


```
