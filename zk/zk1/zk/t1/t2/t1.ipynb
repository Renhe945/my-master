```python
# 1.有1、2、3、4一共四个数字，能组成多少个互不相同且无重复数字的三位数？都是多少？（3分）\
c1=0
for i in range(1,5) :
    for j in range(1,5):
        for ij in range(1,5):
            if i!=j and j!=ij and ij!=i:
                print(i*100+j*10+ij)
                c1+=1
print(f"一共有{c1}个不重复的数字")
```

```python
# 2.一个整数，它加上100后是一个完全平方数，再加上168又是一个完全平方数，问该数是多少？（3分）
# 遍历可能的整数  
for  i in range(10**6):
    # 检查num + 100是否为完全平方数  
    sprt1=(i+100) ** 0.5
    if int(sprt1) ==sprt1:
         # 如果num + 100是完全平方数，则检查(num + 100) + 168是否为完全平方数 
        sprt2=((i+100)+168)*0.5
        if int(sprt2) == sprt2:
            # 如果两个条件都满足，则打印这个数 
            print(i)
        break# 找到答案后退出循环#写这个是找到一个就退出不写输出所有
```

```python
# 5.求100-200之间，有多少个质数，并将对应的数据输出。（3分）
c5=0# 个数
for i in range(100,201):
    # 假设当前数是质数  
    zs=True
    # 检查从2到i的平方根之间的数是否能整除i  
    # 如果能整除，则i不是质数  
    for j in range(2,i):
        if i % j==0:
            zs = False
            break
    if zs:
        print(i)
        c5+=1
print(c5)   
```

```python
# 7.输入一行字符，分别统计出英文字母、空格、数字和其他字符的个数。（3分）
a7=input("请输入一行字符")
y=0
k=0
s=0
q=0
for i in a7:
    if i.isalpha():#字母
        y += 1
    if i.isspace():#空格
        k += 1 
    if i.isdigit():#数字
        s += 1
    else:
        q+=1
print(f"英文字母有{y}个；空格有{k}个；数字有{s}个；其他字符的个数有{q}个")
```
