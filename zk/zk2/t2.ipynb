```python
from pandas import Series
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as ny
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# 题目二 使用sk-learn的逻辑回归完成以下功能（32分）：
# winequality-red.csv是葡萄酒数据集，其中样本label是quality，quality大于4定义为1(合格),小于等于4定义为0（不合格）。
# 1、读取葡萄酒的数据文件winequality-red.csv。（4分）
a1=pd.read_csv("E:///python代码/zk/zk2/data/winequality-red.csv")
# 2、打印出数据的信息和基本统计信息。（4分）
print(a1)
# 4、将数据集自主划分为训练集和测试集。（4分）
x=a1.iloc[:,:-1]
y=a1['quality']
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=33)
# 5、创建逻辑回归的模型。（4分）
lj=LogisticRegression()
# # 6、使用训练集训练逻辑回归模型。（4分）
lj.fit(X_train,y_train)
# # 7、将预测结果打印到控制台。（4分）
yc=lj.predict(X_test)
print("题7",yc)
# 8、使用准确率、精确率和召回率评估模型。（4分）
```
