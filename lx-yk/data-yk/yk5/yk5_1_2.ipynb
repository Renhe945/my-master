```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.svm import SVC
# 题目二、按要求完成下面的各项需求。（30分）
#  鸢尾花.csv 参数依次说明：
# 特征： 花萼长度（Sepal Length）、花萼宽度（Sepal Width）、花瓣长度（Petal Length）
# 花瓣宽度（Petal Width）四个属性; 标签0、1、2分别表示山鸢尾（Setosa）、变色鸢尾（Versicolor）、维吉尼亚鸢尾（Virginical）
# 1．读取鸢尾花.csv数据集, 显示二维数据的统计信息，输出总数、平均值、标准差、最小值等信息。（6分）
import pandas as pd
import  numpy as np
a1=pd.read_csv("E:\\python代码\\lx-yk\\data-yk\\yk5\\鸢尾花.csv")
print(a1.describe())
# 2.划分训练集和测试集。(6分)
x=a1.iloc[:,:-1]
y=a1['Label']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
# 3.创建SVM模型进行预测分类, 分别使用线性,多项式, 高斯核对数据创建模型，惩罚系数为1。（6分）
# 线性
svm_linear  = SVC(kernel='linear',C=1)
svm_linear.fit(x_train,y_train)
linear_pred = svm_linear.predict(x_test)
# 多项式
svm_poly  = SVC(kernel='poly',C=1)
svm_poly.fit(x_train,y_train)
poly_pred = svm_poly.predict(x_test)

# 高斯核                    惩罚系数
svm_rbf  = SVC(kernel='rbf',C=1)
svm_rbf.fit(x_train,y_train)
rfb_pred = svm_rbf.predict(x_test)
# 4.使用混淆矩阵对结果进行进行模型评估。(5分)
# 线性
cm = confusion_matrix(y_test,linear_pred)
cm_display = ConfusionMatrixDisplay(cm).plot()
```

```python
# 多项式
cm1 = confusion_matrix(y_test,poly_pred)
cm_display = ConfusionMatrixDisplay(cm1).plot()
```

```python
# 高斯核    
cm2 = confusion_matrix(y_test,rfb_pred)
cm_display = ConfusionMatrixDisplay(cm2).plot()
```

```python
# # 5.将分类结果分别实现可视化。（5分）
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

import matplotlib.pyplot as plt  
  
# 原始数据的可视化  
plt.figure(figsize=(12, 6))  
  
plt.subplot(1, 3, 1)  
plt.scatter(a1['Petal Length'], a1['Petal Width'], c=a1['Label'], cmap='viridis', marker='o', edgecolor='k')  
plt.title('原始数据')  
plt.xlabel('花瓣长度')  
plt.ylabel('花瓣宽度')  
  
# 线性核SVM分类结果的可视化  
plt.subplot(1, 3, 2)  
plt.scatter(x_test['Petal Length'], x_test['Petal Width'], c=linear_pred, cmap='viridis', marker='o', edgecolor='k')  
plt.title('线性核支持向量机')  
plt.xlabel('花瓣长度')  
plt.ylabel('花瓣宽度')  
  
# 多项式核SVM分类结果的可视化  
plt.subplot(1, 3, 3)  
plt.scatter(x_test['Petal Length'], x_test['Petal Width'], c=poly_pred, cmap='viridis', marker='o', edgecolor='k')  
plt.title('多项式')  
plt.xlabel('花瓣长度')  
plt.ylabel('花瓣宽度')  
  
# 可以在这里添加高斯核SVM分类结果的可视化，但为了避免图表过于拥挤，这里省略  
# 若要添加，可以复制上一个subplot的代码块，并修改相应的预测结果变量为rfb_pred  
  
plt.tight_layout()  
plt.show()

```
