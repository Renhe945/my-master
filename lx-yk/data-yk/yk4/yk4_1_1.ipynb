```python
from json import dump
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
# 题目一、网站点击预测（35分）
# 使用广告点击数据集（网站点击预测.csv），构建分类模型，并且调整超参数获取最佳模型，预测点击。字段如下：
# Daily Time Spent on Site是指用户在网站上的每日时间跨度；
# 年龄是指用户的年龄；
# Area Income指用户所在地区的平均收入；
# Daily Internet Usage是指用户每天的互联网使用情况；
# 广告主题行是指广告的标题；
# 城市是指用户所在的城市；
# 性别是指用户的性别；
# 国家是指用户所在的国家；
# 时间戳是指用户访问网站的时间；
# 如果用户点击了广告，Clicked on Ad表示 1，否则为 0；
# 1.获取数据，将字符类型的特征转换为数字型，将缺失值做均值处理（5分）
import  pandas as pd
import numpy as py
a1=pd.read_csv("E:///python代码/lx-yk/data-yk/yk4/网站点击预测.csv")
# 类型的特征
x=a1.iloc[:,:-1]
y=a1['Clicked on Ad']
# 转换为数字型
a1.info()#查看字段类型
for i in a1.columns:
    if a1[i].dtype=='object':
        a1[i]= pd.factorize(a1[i])[0]
a1
# 将缺失值做均值处理
from sklearn.preprocessing import LabelEncoder
q1={}
for i in ['Ad Topic Line', 'City', 'Gender', 'Country', 'Timestamp']:
    q1[i]=LabelEncoder()
    a1[i]=q1[i].fit_transform(a1[i])
a1
#处理数据
cl=SimpleImputer(strategy='mean')
a1=pd.DataFrame(cl.fit_transform(a1),columns=a1.columns)
a1
# 2.特征选择:选择你认为对模型不重要的特征，然后删除（2分)
x=a1.iloc[:,:-1]
y=a1.iloc[:,1]

# 3.自主划分测试集与训练集，进行算法选型。（5分）
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# 4.至少使用2种算法进行数据建模，并设置合理的参数（5分）
#随机森林建模
model_rf = RandomForestClassifier()
model_rf.fit(x_train,y_train)
yc_tree=model_rf.predict(x_test)
print(yc_tree)

#逻辑回归建模
model_lr = LogisticRegression(C=1.0, max_iter=100, n_jobs=1)
model_lr.fit(x_train,y_train)
yc_lj=model_lr.predict(x_test)
print(yc_lj)
# 5.选择合适评估指标评估模型。（5分）
from sklearn.metrics import accuracy_score
ac1=accuracy_score(y_test,yc_tree)
ac2=accuracy_score(y_test,yc_lj)
from sklearn.metrics import recall_score
r1=recall_score(y_test,yc_tree,average='macro')
r2=recall_score(y_test,yc_lj,average='macro')
from sklearn.metrics import f1_score
f1=f1_score(y_test,yc_tree,average='macro')
f2=f1_score(y_test,yc_lj,average='macro')

# 6.使用合适超参进行微调调整超参数，获取最佳模型。（3分）
cs_tree={'max_depth':[50,100], 'min_samples_split':[2,3,4], 'min_samples_leaf':[1,2,3]}
cs_lr ={'fit_intercept': [True, False], 'max_iter':[10, 30, 40], 'n_jobs':[1,2]}
from sklearn.model_selection import GridSearchCV
cc1=GridSearchCV(estimator=model_rf,param_grid=cs_tree,cv=5)
cc2=GridSearchCV(estimator=model_lr,param_grid=cs_lr ,cv=5)
print(cc1,"=============",cc2)
# 7.使用网格搜索选出最优的模型参数必须包含K折交叉验证，对模型进行预测（5分
cc1.fit(x_train,y_train)
zy7_1=cc1.best_estimator_
print("随机森林建模-最优的模型参数",zy7_1)
yc_7_1=zy7_1.predict(x_test)
ac1=accuracy_score(y_test,yc_7_1)



cc2.fit(x_train,y_train)
zy7_2=cc2.best_estimator_
print("逻辑回归建模-最优的模型参数",zy7_2)
yc_7_2=zy7_2.predict(x_test)
ac1=accuracy_score(y_test,yc_7_2)
# K折交叉验证


# 8.输出最佳模型和最佳参数输出最优模型的权重系数，保存模型到本地（5分）
print("随机森林建模-最优的模型参数",zy7_1)
print("逻辑回归建模-最优的模型参数",zy7_2)
a = pd.DataFrame({'features': x.columns, '权重系数': zy7_1.feature_importances_})

print(a)
dump(zy7_1,'yk4_1.pkl')
dump(zy7_2,'yk4_2.pkl')


```
