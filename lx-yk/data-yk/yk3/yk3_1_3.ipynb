```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import jieba

# 1.读取数据
a1 = pd.read_excel("E:///python代码/lx-yk/data-yk/yk3/新闻.xlsx")

# 2.遍历来进行所有标题的分词
bl = a1['标题']
words = []
for i in bl:
    word = jieba.cut(i)
    a2 = " ".join(word)
    words.append(a2)

# 4.建立词频矩阵
jz = CountVectorizer()
x = jz.fit_transform(words)
x_array = x.toarray()
# 5.查看词袋和对应的顺序(3分)
print("查看词袋\n",jz.get_feature_names_out())
print("对应的顺序\n",jz.vocabulary_)

# 6.将之前所有的新闻标题进行文本向量化
jz1 = CountVectorizer()
x1 = jz1.fit_transform(words)
x1_array = x1.toarray()
# # 7.查看所有新闻标题的词袋
print('7提', jz1.get_feature_names_out())
# # 8.文本向量化
t8 = jz1.transform(bl)
print('8提', t8.toarray())
# # 9.查看词袋的第一种方式
print('9提', jz1.get_feature_names_out())

# # 10.查看词袋的第二种方式
print('10提', jz1.vocabulary_)
# 
# # 11.将之前所有的新闻标题进行文本向量化并通过pandas展示
a11 = pd.DataFrame(x1_array, columns=jz1.get_feature_names_out())
a11.head()

# 12.自主选择算法进行聚类分群
print("12.自主选择算法进行聚类分群\n")
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=123)
kmeans.fit(x1_array)
a1['聚类标签']=kmeans.labels_
print(a1[['标题', '聚类标签']])


```
