````python
# **题目2：性别统计与可视化**（3分）
# 统计不同性别的患者数量，并使用Matplotlib绘制饼图展示。
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['Simhei']
plt.rcParams['axes.unicode_minus'] = False
xb=a1['性别'].value_counts()#计算男和女各自出现的次数。
## 从xb中获取索引，即性别值（'男'和'女'），作为饼图的标签
labels = xb.index #Index(['男', '女'], dtype='object', name='性别')
# 从xb中获取数据值，即性别出现的次数，作为饼图的大小  
sizes = xb.values#array([8, 7], dtype=int64)
# 创建一个新的图形和子图 
plt.subplots()
plt.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal') 
plt.title("性别分布")
plt.show()
sizes
````

````python

# **题目3：年龄分布与可视化**（3分）
# 使用Matplotlib绘制直方图，展示患者年龄分布。、、
plt.hist(a1['年龄'],bins=10)
plt.title('年龄分布与可视化')
plt.xlabel('年龄')
plt.ylabel('频数')
plt.show()
````

````python

# **题目4：医疗费用分布与可视化**（5分）
# 使用Matplotlib绘制箱线图，展示医疗费用的分布情况。
data= a1['医疗费用']
plt.boxplot(data)
plt.title("医疗费用分布与可视化")
plt.ylabel('医疗费用')
plt.show()
````

````python
from matplotlib.pyplot import figure

# **题目5：不同性别医疗费用对比**（5分）
# 使用Pandas分组计算不同性别患者的平均医疗费用，并使用Matplotlib绘制柱状图展示。

# 使用Pandas分组计算不同性别患者的平均医疗费用  
fz = a1.groupby('性别')['医疗费用'].mean()  
  
# 绘制柱状图展示  
fz.plot(kind='bar')  
plt.title('不同性别患者的平均医疗费用')  
plt.xlabel('性别')  
plt.ylabel('平均医疗费用')  
plt.show()


````

````python
# **分析业务二：诊断分析与治疗方案建议**
# **题目6：疾病诊断分布与可视化**（5分）
# 统计不同年龄段和诊断疾病的数量，并使用Matplotlib绘制柱状图展示并形成结论。
age_bins = [0, 20, 40, 60, 80, 100]  
age_labels = ['0-20', '21-40', '41-60', '61-80', '81+']  
a1['年龄组'] = pd.cut(a1['年龄'], bins=age_bins, labels=age_labels)  #
a2=a1.groupby(['年龄组', '诊断'],observed=False).size().unstack(fill_value=0) 
# 绘制柱状图  
a2.plot(kind='bar', stacked=True)  
plt.title('不同年龄段和诊断疾病的数量')  
plt.xlabel('年龄组')  
plt.ylabel('疾病数量')  
plt.legend(title='诊断')  
plt.show()  
````

````python
# **题目7：不同疾病平均医疗费用对比**（5分）
# 使用Pandas分组计算不同诊断疾病的平均医疗费用，并使用Matplotlib绘制柱状图展示并形成结论。
t7=a1.groupby('诊断')['医疗费用'].mean()
t7.plot(kind='bar')
plt.title('不同疾病平均医疗费用对比')
plt.xlabel('诊断疾病')
plt.ylabel('平均医疗费用')
plt.show()
#形成结论
````
