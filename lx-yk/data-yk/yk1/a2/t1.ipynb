```python
# 题目一、质量分类（35分）。
# 红葡萄酒数据集的样本是通过在葡萄牙北部的"Vinho Verde"产区收集的。红葡萄酒数据集是一个多变量数据集，其中的特征变量是描述红葡萄酒化学特性的数值。这些特征变量的值可以用来推断红葡萄酒的质量评分。样本label是quality，quality大于4定义为1(合格),小于等于4定义为0（不合格）。
# 获取数据：winequality-red.csv
# 1、数据探索分析：
import pandas as pd
import numpy as np
# 1）查看前5行（2分）
a1=pd.read_csv("E:\\python代码\\lx-yk\\data-yk\\yk1\\winequality-red.csv")
print('查看前5行',a1.head(5))
# 2）查看数据信息（2分）
print('查看数据信息',a1.info)
# 3）查看基本统计信息（2分）
print('查看基本统计信息',a1.density)
# 4）查看数据的行数和列数（2分）
print('查看数据的行数和列数',a1.shape)
```

```python
from sklearn.metrics import accuracy_score, precision_recall_curve, f1_score, precision_score, recall_score
from sklearn.model_selection import train_test_split

# 2、自主划分训练集、验证集或测试集(2分)
# 处理数据
# quality大于4定义为1(合格),小于等于4定义为0（不合格）。
x=a1.iloc[:,:-1]
y=a1['quality']
y=np.where(y>4,1,0)
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42)
# 3、自主选择两种算法进行数据建模。（2分）
from sklearn.linear_model import LogisticRegression
lj = LogisticRegression().fit(X_train, y_train)

from sklearn.tree import DecisionTreeClassifier
sj = DecisionTreeClassifier().fit(X_train, y_train)
# 4、在训练集和测试集上测试该模型的准确性，并展示出该模型的训练准确率和测试准确率（5分）
# 在训练集中测试准确性       
yc1_xl=lj.predict(X_train)
yc2_xl=sj.predict(X_train)
# 测试集上测试该模型的

yc1_cs=lj.predict(X_test)
yc2_cs=sj.predict(X_test)


# 训练准确率
print('逻辑回归训练准确率',accuracy_score(y_train,yc1_xl))
print('决策树  训练准确率',accuracy_score(y_train,yc2_xl))
# 测试准确率
print('逻辑回归测试准确率',accuracy_score(y_test,yc1_cs))
print('决策树 测试准确率',accuracy_score(y_test,yc2_cs))


# 5、使用K折交叉验证进行参数调优，直到获得最优模型（5分）
from sklearn.model_selection import KFold
k=KFold(n_splits=5, random_state=None, shuffle=True)
lj_ty={'C':[0.1,1.0,10],'max_iter':[2,3,4]}
sj_ty={'max_depth':[2,3,4],'min_samples_split':[2,3,4]}

from sklearn.model_selection import GridSearchCV
#网络调优
clf_lj = GridSearchCV(lj,lj_ty,cv=k)
# 最优模型
zy_mx1=clf_lj.fit(X_train, y_train).best_estimator_
print('最优模型',zy_mx1)

clf_sj = GridSearchCV(sj,sj_ty,cv=k)
zy_mx2=clf_sj.fit(X_train, y_train).best_estimator_
print('最优模型',zy_mx2)

# 6、模型评估：使用你认为合适的指标评估模型。（至少两种）（5分）
jq=precision_score(y_test, zy_mx1.predict(X_test))
zh=recall_score(y_test, zy_mx1.predict(X_test))
zq=accuracy_score(y_test, zy_mx1.predict(X_test))
f1=f1_score(y_test, zy_mx1.predict(X_test))
print(f"逻辑回归 精确率{jq}\n 召回率{zh}\n 准确率{zq}\n f1{f1}\n")


jq1=precision_score(y_test, zy_mx2.predict(X_test))
zh1=recall_score(y_test, zy_mx2.predict(X_test))
zq1=accuracy_score(y_test, zy_mx2.predict(X_test))
f11=f1_score(y_test, zy_mx2.predict(X_test))
print(f"决策树 精确率{jq1}\n 召回率{zh1}\n 准确率{zq1}\n f1{f11}\n")
```

```python
# 7、绘制出模型的AUC曲线下面积，并输出最优模型的权重系数，并对样本进行预测。（8分）
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif']=['Simhei']
plt.rcParams['axes.unicode_minus']=False

fpr, tpr, thresholds = metrics.roc_curve(y_test, zy_mx2)
t7=metrics.auc(fpr, tpr)

t7.plot()
plt.show()









```
