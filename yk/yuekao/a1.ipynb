```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder, Binarizer
from sklearn.model_selection import train_test_split

'''题目一、 森林火灾燃烧面积统计分析。（44分）
附件ForestFires.csv中，是葡萄牙东北部地区森林火灾的燃烧面积数据，请根据特征预测出燃烧面积（area）的数据。
X	Montesinho 公园地图内的 x 轴空间坐标：1 到 9
Y	Montesinho 公园地图中的 y 轴空间坐标：2 到 9
month	一年中的月份：“Jan”到“Dec”
day	星期几：“mon”到“sun”

FFMC	FWI 系统的 FFMC 指数：18.7 至 96.20
DMC	FWI 系统的 DMC 指数：1.1 至 291.3
DC	FWI 系统的直流指数：7.9 至 860.6
ISI	FWI 系统的 ISI 指数：0.0 至 56.10
temp	温度：2.2 至 33.30
RH	相对湿度：15.0 至 100
wind	风速：0.40 至 9.40
rain	室外降雨量：0.0 至 6.4
area	森林的燃烧面积：0.00 到 1090.84
'''
import pandas as pd
import numpy as py

# 1、获取数据：ForestFires.csv，并查看数据前10行。(2分)
a1=pd.read_csv('E:\\python代码\\yk\\yuekao\\ForestFires.csv')
# 查看数据前10行。
print("并查看数据前10行。\n",a1.head(10))
```

```python
# 2、查看数据的基本统计信息和行列数。(2分)
print('2、查看数据的基本统计信息\n',a1.backfill)
print("查看行列",a1.shape)
```

```python
# 3、如果数据中有缺失值，则将该条数据删除。(2分)
#查看是否有缺失值
a1.isnull().any()
#删除缺失值
a1=pd.DataFrame(a1).dropna()
a1.isnull().sum()
```

```python
# 4、检查是否有重复数据，如果有则删除。(2分)\
# 检查是否有重复数据
a1.duplicated().any()
#删除重复值
a1=a1.drop_duplicates()
a1.duplicated().any()
```

```python
# # 5、对“DC”字段的数据进行归一化处理。(4分)、
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
a2=a1['DC'].values
a3=pd.DataFrame(a2)
a1['DC']=scaler.fit_transform(a3)
```

```python
# # 6、将“month”和“day”字段的值转换成数值格式。(4分)
#查看数据值格式。
a1['month']
le = LabelEncoder()
a1['month']=le.fit(a1['month']).transform(a1['month'])
print('转换后的month类型',a1['month'].reset_index)
le1 = LabelEncoder()
a1['day']=le1.fit(a1['day']).transform(a1['day'])

print('转换后的day类型',a1['day'].reset_index)
```

```python
# # 7、获取特征的Pearson相关系数矩阵，并选择最优的8个特征。(4分)
x=a1[["FFMC","DMC","DC","ISI","temp","RH","wind","rain"]]
y=a1['area']
print(x,'\n',y)
```

```python
# # # 8、按照8:2将数据集划分成测试集与训练集。(4分)
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=8)
X_train
```

```python
from sklearn.linear_model import Ridge, Lasso
# 9、创建线性回归的L1正则化模型和L2正则化模型。（4分）
from sklearn.linear_model import ElasticNet
reg_l1 = Ridge(alpha=1.0).fit(X_train, y_train)
reg_l2 = Lasso(alpha=1.0).fit(X_train, y_train)

t9_yc1=reg_l1.predict(X_test)
t9_yc2=reg_l2.predict(X_test)
```

```python
# 10、使用网格搜索对以上模型进行参数调优，最少设置两个超参数。（4分）
from sklearn.model_selection import GridSearchCV
cs10_1={"alpha":[1.0,2.0,10.0],"max_iter":[100,200,500]}
cs10_2={"alpha":[1.0,2.0,10.0],"max_iter":[100,200,500]}
clf1 = GridSearchCV(reg_l1, cs10_1,cv=5)
clf1.fit(X_train, y_train)
print("l1最优参数",clf1.best_params_)

clf2 = GridSearchCV(reg_l2, cs10_2,cv=5)
clf2.fit(X_train, y_train)
print("l2最优参数",clf2.best_params_)
```

```python
# 11、分别获得最优模型，并打印对应的截距和斜率。(4分)
t11_zy1=clf1.best_estimator_
print('最优模型1',t11_zy1)
print('对应的截距',t11_zy1.intercept_)
print('斜率',t11_zy1.coef_)

print("--------------------------------------------------")

t11_zy2=clf1.best_estimator_
print('最优模型2',t11_zy2)

print('对应的截距',t11_zy2.intercept_)
print('斜率',t11_zy2.coef_)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# 12、使用至少两种指标对模型进行评估。(4分)
t12_1=t11_zy1.predict(X_test)
r2_score(y_test, t12_1)
print('R2',r2_score(y_test, t12_1))
print('均方误差回归损失',mean_squared_error(y_test, t12_1))




t12_2=t11_zy2.predict(X_test)
r2_score(y_test, t12_2)
print('R2',r2_score(y_test, t12_2))
mean_squared_error(y_test, t12_2)
print('均方误差回归损失',mean_squared_error(y_test, t12_2))



```
