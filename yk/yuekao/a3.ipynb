```python
# 题目三 、实现雷达回波聚类统计分析（28分）
# Ionosphere.csv数据集中存放的是电离层雷达回波的数据。使用k-means对其进行聚类。

import pandas as pd
import numpy as py
a1=pd.read_csv('E:\\python代码\\yk\\yuekao\\Ionosphere.csv')
# 1、读取数据文件并查看数据基本统计信息。(2分)
print("读取数据\n",a1)
print('查看基本统计数据信息\n',a1.describe())

# 2、创建聚类k-means模型，设置k=2。（2分）
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0, n_init="auto").fit(a1)
kmeans
```

```python
# 3、打印出聚类中心点的坐标、各数据点所属聚类中心点的位置，以及对应的代价值。(4分)
print('聚类中心点的坐标',kmeans.cluster_centers_)
kmeans.labels_
gg=kmeans.cluster_centers_
print("各数据点所属聚类中心点的位置\n",gg)
t3=kmeans.inertia_

print('对应的代价值\n',t3)
```
