```python

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as py
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor

# 1.随机产生400个0-5之间的随机数。(10分)
a1=py.random.randint(0,6,400)
# a1
# 2.使用上面的随机数计算sin(x)值。(10分)
a2 =py.sin(a1)
a2
# 3.划分训练集和测试集。（10分）
x=a1.reshape(-1,1)
y=a2
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=0)
# 4.调用决策树库函数创建决策树模型。（10分）
trees1=DecisionTreeRegressor(max_depth=2)
# 5.使用训练集训练模型，深度为2。（10分）
trees1.fit(x_train,y_train)
# 6.计算深度为2的准确率。（10分）
yc1=trees1.predict(x_test)
zql1=mean_squared_error(yc1,y_test)
print("准确率深度为2\n",zql1)
# 7.使用训练集训练新的决策树模型，深度为5。（10分）
trees2=DecisionTreeRegressor(max_depth=5)
trees2.fit(x_train,y_train)
# 8.计算深度为5的准确率。（10分）
yc2=trees2.predict(x_test)
zql2=mean_squared_error(yc2,y_test)
print("准确率深度为5\n",zql2)
```

```python
# 9.画出深度=2时的决策树曲线.(10分)
import matplotlib.pyplot as plt
x=py.linspace(0, 2, 100).reshape(-1,1)
plt.figure(figsize=(10, 6))#, layout='constrained'
plt.subplot(1,2,1)
plt.plot(x_test,y_test,label='linear')
from sklearn import tree
plt.plot(x,trees1.predict(x),label = "Tree 5",color = 'y')
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Tree Depth = 5")
plt.legend()
plt.show()

```
