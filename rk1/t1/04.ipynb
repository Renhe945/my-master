```python
# Python使用类库连接mysql，并使用SQL进行数据库增删改查操作：

# **(1)**
# 题目：编写一个SQL语句，在名为"users"的表中插入一条记录，姓名为"Alice"，年龄为25岁。
import  pymysql
a1=pymysql.connect(
    host="localhost",
    database="a",
    user="root",
    password="1234",
    port=3306
)
#游标
cursor=a1.cursor()
sql="insert into user(name,age) values('aa',11)"
cursor.execute(sql)#执行
a1.commit()#提交

sql = "select * from user"
# 执行sql
cursor.execute(sql)
result = cursor.fetchall()#查询结果的所有行
for res in result:
    print(res)#('Alice', 25)('aa', 11)
```
